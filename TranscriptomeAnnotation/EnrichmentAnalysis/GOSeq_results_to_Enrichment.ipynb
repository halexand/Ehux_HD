```python
import pandas as pd
import numpy as np
import glob
```

```python
for filename in glob.iglob('*[RVX]*.Go_Wall.txt'):
    print filename
    GOwall = pd.read_table(filename)
    GOwall=GOwall.set_index('category') 
    GOUP=(GOwall.over_represented_pvalue<0.05)
    GODN=(GOwall.under_represented_pvalue<0.05)
    UpGOWall=GOwall[GOUP]
    DnGOWall=GOwall[GODN]
    print(len(UpGOWall), len(DnGOWall))
    UpGOWall[['over_represented_pvalue']].to_csv(filename[0:4]+'overrep.txt', header=False, sep='\t')
    DnGOWall[['under_represented_pvalue']].to_csv(filename[0:4]+'underrep.txt', header=False,sep='\t')
```

```python
GOUP=(GOwall.over_represented_pvalue<0.05)
GODN=(GOwall.under_represented_pvalue<0.05)
SigGOWall=GOwall[GOUP | GODN]
SigGOWall=SigGOWall.set_index('category')
SigGOWall
```

```python
SigGOWall['p.val']=np.nan
SigGOWall['Phenotype']=np.nan

for i in SigGOWall.index:
    if SigGOWall.loc[i, 'over_represented_pvalue'] < 0.05:
        SigGOWall.loc[i, 'pvalue']=SigGOWall.loc[i, 'over_represented_pvalue']
        SigGOWall.loc[i,'Phenotype']= '+1'
    else: 
        SigGOWall.loc[i, 'pvalue']=SigGOWall.loc[i, 'under_represented_pvalue']
        SigGOWall.loc[i,'Phenotype']= '-1'

SigGOWall['FDR']=np.nan
        
```

```python
OutputPD=SigGOWall[['term', 'pvalue', 'FDR', 'Phenotype']]
```

```python
OutputPD.to_csv('QvR.Generic.txt', sep ='\t', na_rep='')
```

```python
f=open('out.test', 'w')
f.write("GO.ID\tDescription\tp.Val\t\tPhenotype\n")
for i in OutputPD.index:
    f.write(i)
    f.write('\t')
    f.write(str(OutputPD.loc[i,'term']))
    f.write('\t')
    f.write(str(OutputPD.loc[i,'pvalue']))
    f.write('\t')
    f.write('\t')
    f.write(OutputPD.loc[i,'Phenotype'])
    f.write('\n')
```
